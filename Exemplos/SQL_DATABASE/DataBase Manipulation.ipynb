```python
import sqlite3
import os
import pandas as pd
```

```python
conn = sqlite3.connect('nanobio_data.db')
c = conn.cursor()

def tableCreate_ligand():
    c.execute('CREATE TABLE Ligand (ID INTEGER PRIMARY KEY AUTOINCREMENT, protein_name TEXT, description TEXT, protein_loc TEXT)')
def tableCreate_protein():
    c.execute('CREATE TABLE Protein (ID INTEGER PRIMARY KEY AUTOINCREMENT, ligand_name TEXT, description TEXT, ligand_loc TEXT)')
def tableCreate_simvina():
    c.execute('CREATE TABLE SimVina (ID INTEGER PRIMARY KEY AUTOINCREMENT, ligand_id INT, protein_id INT, ligand_name TEXT, protein_name TEXT, count INT, x REAL, y REAL, z REAL, energy REAL, energypoints REAL, arc_name TEXT)')
```

```python
tableCreate_ligand(), tableCreate_protein(), tableCreate_simvina()
```

```python
def dataEntry_ligand(x):
    c.execute(f'INSERT INTO Ligand (protein_name, description, protein_loc) VALUES ("{x[0]}", "{x[1]}", "{x[2]}");')
    conn.commit()
```

```python
def dataEntry_protein(x):
    c.execute(f'INSERT INTO Protein (ligand_name, description, ligand_loc) VALUES ("{x[0]}", "{x[1]}", "{x[2]}");')
    conn.commit()
```

## Treating Ligand table

```python
path_r = r'data\Ligand'
path = os.path.join(os.getcwd(), path_r)
```

```python
names = []
for file in os.listdir(path):
    names.append(file)

for name in names:
    dataEntry_ligand([name, 'Ligand ' + str(name) + ' with only polar hydrogen', os.path.join(path_r, name)])
```

```python
df_ligand = pd.read_sql_query("SELECT * from Ligand", conn, index_col='ID')
df_ligand
```

## Treating Protein table

```python
path_r = r'data\Protein'
path = os.path.join(os.getcwd(), path_r)
```

```python
names = []
for file in os.listdir(path):
    names.append(file)

for name in names:
    dataEntry_protein([name, 'Protein ' + str(name.split('.')[0]) + ' with only polar hidrogen', os.path.join(path_r, name)])
```

```python
df_protein = pd.read_sql_query("SELECT * from Protein", conn, index_col='ID')
df_protein
```
